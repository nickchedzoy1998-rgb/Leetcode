```python
roman_numerals = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000}
```

```python
def roman_to_integer(numerals):
    total = 0
    numerals = list(numerals)

    for i, numeral in enumerate(numerals):
        current_value = roman_numerals[numeral]

        # If there *is* a next numeral, get its value
        if i + 1 < len(numerals):
            next_value = roman_numerals[numerals[i+1]]
        else:
            next_value = 0  # no next numeral

        # Subtraction rule
        if current_value < next_value:
            total -= current_value
        else:
            total += current_value

    return total
```

```python
roman_to_integer('III')
```

```python
roman_to_integer('LVIII')
```

```python
roman_to_integer('MCMXCIV')
```
