```python
#Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

#You may assume that each input would have exactly one solution, and you may not use the same element twice.

#You can return the answer in any order.
```

```python
# Initial Submission

def find_indices(nums, target):
    complements_needed = []
    indices = []
    unique_check = {}

    for num in nums:
        if num not in unique_check.keys():
            unique_check[num] = 1
        else:
            unique_check[num] += 1

    for idx, num in enumerate(nums):
        complement = target - num
        complements_needed.append(complement)

        if len(indices) < 2:
            if complement in nums:

                if complement != num:
                    indices.append(idx)
                    indices.append(nums.index(complement))
            
                if complement == num:
                    if unique_check[num] > 1:
                        indices.append(idx)
                        nums.remove(num)
                        indice_2 = nums.index(complement)
                        indice_2 = indice_2 + 1
                        indices.append(indice_2)
                            
    return indices
```

```python
nums = [2,7,11,15]
target = 9

# Output [0,1]
find_indices(nums, target)
```

```python
nums = [3,2,4]
target = 6

# Output [1,2]
find_indices(nums, target)
```

```python
nums = [3,3]
target = 6

# Output [0,1]
find_indices(nums, target)
```

```python
# Submission 2
# Optimise for 0(n^2)

def find_indices(nums, target):
    indices = []

    for idx, num in enumerate(nums):
        complement = target - num

        if complement not in nums:
            continue

        else:
            complement_idx = nums.index(complement)

            if idx != complement_idx:
                indices.append(idx)
                indices.append(complement_idx)
                break
                

            else:
                # Check if exists in list twice
                filtered_nums = nums
                filtered_nums = nums.count(num)
                if filtered_nums < 2:
                    filtered_nums = []
                    continue
                else:
                    adjusted_nums = list(nums)
                    adjusted_nums.remove(num)
                    complement_idx = adjusted_nums.index(complement)
                    complement_idx = complement_idx + 1
                    indices.append(idx)
                    indices.append(complement_idx)
                    adjusted_nums = list(nums)
                    if len(indices) == 2:
                        break

    return indices
```
