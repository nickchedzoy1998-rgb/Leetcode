```python
def longestCommonPrefix(strs):
    no_words = len(strs)

    shortest_word_len = min(len(word) for word in strs)

    prefix = []

    for i in range(shortest_word_len):
        char = list(strs[0])[i]
        mismatch = False

        for word in strs:
            if word[i] != char:
                mismatch = True
                break

        if mismatch:
            break
        else:
            prefix.append(char)

    if len(prefix) == 0:
        return("")
    
    else:
        prefix = "".join(prefix)
        return prefix
```

```python
str(longestCommonPrefix(["flower","flow","flight"]))
```

```python
longestCommonPrefix(["dog","racecar","car"])
```
